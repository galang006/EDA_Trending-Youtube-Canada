```python
import pandas as pd

df = pd.read_csv('CAvideos.csv')

df.head()
```

```python
data = df.drop(columns=['video_id', 'category_id', 'thumbnail_link',	'comments_disabled'	, 'ratings_disabled'	, 'video_error_or_removed'	])
data
```

```python
data.info()
```

```python
data.describe()
```

```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
# Buat plot histogram dengan sns.histplot
sns.set(style="ticks")
g = sns.histplot(data = data, x="views", kde=False, color="#268bd2", bins=47 , alpha=1)

# Atur label dan judul plot
g.set(xlabel="Views", ylabel="No. of videos")
plt.title("Histogram of Views")
```

```python
# Buat plot histogram dengan sns.histplot
sns.set(style="ticks")
g = sns.histplot(data = data, x="views", kde=False, color="#268bd2", bins=47, alpha=1)

g.set_xlim(0, 10000000)
# Atur label dan judul plot
g.set(xlabel="Views", ylabel="No. of videos")
plt.title("Histogram of Views")
```

Now we see that the majority of trending videos have 5 million views or less. Let's see the exact percentage of videos less than 5 million views

```python
data[data['views'] < 5e6]['views'].count() / data['views'].count() * 100
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Gantilah 'df' dengan DataFrame yang berisi dataset Anda
# Pastikan Anda sudah mengimpor pandas
# Sebagai contoh, saya akan membuat DataFrame dengan data video dan jumlah viewnya:

# data = {
#     'Views': [100, 100, 200, 300, 400, 500, 500]
# }
df = data

# Menghitung frekuensi kemunculan setiap nilai
value_counts = df['views'].value_counts()

# Membuat histogram untuk frekuensi kemunculan data
plt.figure(figsize=(10, 6))
plt.hist(df['views'], bins=len(value_counts), edgecolor='k', alpha=0.7)

# Menyeting label
plt.xlabel('Number of Views')
plt.ylabel('Frequency')

# Menampilkan plot
plt.show()
```
